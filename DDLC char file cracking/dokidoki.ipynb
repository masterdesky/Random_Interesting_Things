### Decode binary code on image

```python
monika_binary = plt.imread(monika_binary_image)[:,:,:3]
```

```python
monika_bitcoded = np.zeros((monika_binary.shape[0], monika_binary.shape[1]))
for i, row in enumerate(monika_binary):
    for j, cell in enumerate(row):
        monika_bitcoded[i][j] = np.mean(cell)
```

```python
nrows = 1
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*11,nrows*11))

axes.imshow(monika_bitcoded,
            cmap='Greys_r')

plt.show()
```

```python
bitcode_str = '0b'
for i in monika_bitcoded.flatten():
    bitcode_str += str(int(i))
```

```python
bit_int = int(bitcode_str, 2)
bitcode_decoded = bit_int.to_bytes((bit_int.bit_length() + 7) // 8, 'big').decode()
bitcode_stipped = bitcode_decoded.split('=')[0] + '='
monika_decoded = base64.b64decode(bitcode_stipped).decode('utf8')
```

```python
print(monika_decoded)
```

```python
monika_text_file = files + 'monika - text.txt'
```

```python
with open(monika_text_file, 'w+') as f:
    f.write(monika_decoded)
```

```python
fig = mlab.figure(size=(800,800))

image_file = natsuki_negative_image

# load and map the texture
img = tvtk.PNGReader()
img.file_name = image_file
texture = tvtk.Texture(input_connection=img.output_port, interpolate=1)
# (interpolate for a less raster appearance when zoomed in)

# create the cone source with a given height, radius and angular resolution
cone = tvtk.ConeSource(height=10.0, radius=10.0, resolution=180)

# assemble rest of the pipeline, assign texture    
cone_mapper = tvtk.PolyDataMapper(input_connection=cone.output_port)
cone_actor = tvtk.Actor(mapper=cone_mapper, texture=texture)

fig.scene.add_actor(cone_actor)

display(fig)
```

```python
fig = mlab.figure(size=(800,800))

image_file = natsuki_negative_image

# load and map the texture
img = tvtk.PNGReader()
img.file_name = image_file
texture = tvtk.Texture(input_connection=img.output_port, interpolate=1)
# (interpolate for a less raster appearance when zoomed in)

# use a TexturedSphereSource, a.k.a. getting our hands dirty
Nrad = 180

# create the sphere source with a given radius and angular resolution
sphere = tvtk.TexturedSphereSource(radius=30, theta_resolution=Nrad,
                                   phi_resolution=Nrad)

# assemble rest of the pipeline, assign texture    
sphere_mapper = tvtk.PolyDataMapper(input_connection=sphere.output_port)
sphere_actor = tvtk.Actor(mapper=sphere_mapper, texture=texture)

fig.scene.add_actor(sphere_actor)

display(fig)
```

```python
shutil.copy(sayori_file, files + 'sayori.ogg')
```

```python
sayori_ogg = files + 'sayori.ogg'
sayori_wav = files + 'sayori.wav'
```

```python
# Read .ogg
sayori_ogg_data, sample_rate = sf.read(sayori_ogg)
# Write .wav for easier handling
sf.write(sayori_wav, sayori_ogg_data, sample_rate)
```

```python
sayori_sound_samples = sayori_ogg_data.shape[0]
sayori_sound_length = sayori_sound_samples/sample_rate
t = np.linspace(0,sayori_sound_length,sayori_sound_samples)
```

```python
nrows = 1
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*20,nrows*10))

axislabelsize = 25
axisticksize = 17

axes.plot(t, sayori_ogg_data)
axes.set_xlabel('Time [s]', fontsize=axislabelsize)
axes.set_ylabel('Amplitude', fontsize=axislabelsize)
axes.tick_params(axis='both', which='major', labelsize=axisticksize)

plt.savefig(files + 'sayori_wave.png')

plt.show()
```

```python
IPython.display.Audio(sayori_wav)
```

```python
f, t, Sxx = signal.spectrogram(sayori_ogg_data, sample_rate)
```

```python
nrows = 1
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*20,nrows*20))

axislabelsize = 25
axisticksize = 15

axes.pcolormesh(t, f, Sxx, cmap='Greys_r')
axes.set_ylabel('Frequency [Hz]', fontsize=axislabelsize)
axes.set_xlabel('Time [sec]', fontsize=axislabelsize)
axes.tick_params(axis='both', which='major', labelsize=axisticksize)
axes.set_ylim(8000,)
axes.axis('off')

plt.savefig(files + 'sayori_spectrogram.png')

plt.show()
```
