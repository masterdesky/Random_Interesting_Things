# Create 3D pictures for Facebook

```python
import os
import numpy as np

import seaborn
import matplotlib.pyplot as plt
```

## Creating doge

```python
figsave_format = 'png'
figsave_dpi = 150

doge_orig = 'src/doge_edit_2.jpg'
doge_edit = 'src/doge_edit.jpg'
doge_shuf = 'src/doge_shuf.jpg'

# Load pictures
image_orig = plt.imread(doge_orig)
image_orig.setflags(write=1)

image_edit = plt.imread(doge_edit)
image_edit.setflags(write=1)

image_shuf = plt.imread(doge_shuf)
image_shuf.setflags(write=1)
```

#### Addig alpha layer

```python
def add_alpha(image_orig, alpha=255):
    # Define RGBA picture size
    alpha_shape = list(image_orig.shape)
    # Adding alpha
    alpha_shape[-1] += 1
    # Creating dummy ndarray for RGBA image
    image_orig_alpha = np.zeros(tuple(alpha_shape), dtype='uint8')

    for i in range(image_orig.shape[0]):
        for j in range(image_orig.shape[1]):
            pixel = image_orig[i,j,:]
            pixel = np.append(arr=pixel, values=alpha)
            image_orig_alpha[i,j,:] = pixel
            
    return(image_orig_alpha)
```

```python
nrows = 1
ncols = 1
inv_figscale = 60
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*image_orig.shape[0]/inv_figscale,
                                                            nrows*image_orig.shape[1]/inv_figscale))

titlesize = 20

axes.imshow(add_alpha(image_orig=image_orig, alpha=150))
axes.imshow(add_alpha(image_orig=image_edit, alpha=150))
axes.set_title('Original image', fontsize=titlesize, fontweight='bold')

plt.show()
```

### Simple transform on original image

```python
image_trans = image_orig.copy()
row_step = 1
col_step = 10
for i in range(image_trans.shape[0]//row_step):
    for j in range(image_trans.shape[1]//col_step):
        image_trans[i*row_step, j*col_step, :] = np.zeros((3,), dtype='uint8')
```

```python
nrows = 1
ncols = 1
inv_figscale = 60
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*image_orig.shape[0]/inv_figscale,
                                                            nrows*image_orig.shape[1]/inv_figscale))

titlesize = 20

axes.imshow(image_trans)
axes.set_title('Transformed image', fontsize=titlesize, fontweight='bold')

plt.imsave(fname='out/doge_lined.' + figsave_format,
           arr = image_trans,
           format=figsave_format,
           dpi=figsave_dpi)
plt.show()
```

### Slice transform

```python
def slice_image(image_orig, row_step=None, col_step=None, step=2, offset=0,
                alpha=255, only_draw=False, inverse=False):

    # Create an empty image
    image_sliced = image_orig.copy()
    if image_sliced.shape != (image_orig.shape[0], image_orig.shape[0], 4):
        image_sliced = add_alpha(image_orig=image_orig, alpha=alpha)

    for idx, row_col in enumerate([row_step, col_step]):
        if row_col is not None and row_col is not 0:
            # Gather pixel indeces
            slices = [i*row_col for i in range(image_sliced.shape[idx]//row_col + 1)]
            
            if inverse:
                slices = slices[::-1]
                for i in slices[:-1]:
                    if idx == 0:
                        shape = image_sliced[i-row_col:i, :, :].shape
                        if only_draw:
                            if (i/row_col+offset)%step == 0:
                                image_sliced[i-row_col:i, :, :] = np.zeros(shape, dtype='uint8')
                        else:
                            if (i/row_col+offset)%step != 0:
                                image_sliced[i-row_col:i, :, :] = np.zeros(shape, dtype='uint8')
                    elif idx == 1:
                        shape = image_sliced[:, i-row_col:i, :].shape
                        if only_draw:
                            if (i/row_col+offset)%step == 0:
                                image_sliced[:, i-row_col:i, :] = np.zeros(shape, dtype='uint8')
                        else:
                            if (i/row_col+offset)%step != 0:
                                image_sliced[:, i-row_col:i, :] = np.zeros(shape, dtype='uint8')
            else:
                for i in slices[:-1]:
                    if idx == 0:
                        shape = image_sliced[i:i+row_col, :, :].shape
                        if only_draw:
                            if (i/row_col+offset)%step == 0:
                                image_sliced[i:i+row_col, :, :] = np.zeros(shape, dtype='uint8')
                        else:
                            if (i/row_col+offset)%step != 0:
                                image_sliced[i:i+row_col, :, :] = np.zeros(shape, dtype='uint8')
                    elif idx == 1:
                        shape = image_sliced[:, i:i+row_col, :].shape
                        if only_draw:
                            if (i/row_col+offset)%step == 0:
                                image_sliced[:, i:i+row_col, :] = np.zeros(shape, dtype='uint8')
                        else:
                            if (i/row_col+offset)%step != 0:
                                image_sliced[:, i:i+row_col, :] = np.zeros(shape, dtype='uint8')

    return(image_sliced)
```

```python
# Hyperparameters
row_step = 20
col_step = None

doge_1 = slice_image(image_orig, row_step=row_step, col_step=col_step, step=3, offset=0, 
                     alpha=255, only_draw=False, inverse=False)
doge_2 = slice_image(image_edit, row_step=row_step, col_step=col_step, step=3, offset=2,
                     alpha=255, only_draw=False, inverse=False)
doge_3 = slice_image(image_shuf, row_step=row_step, col_step=col_step, step=3, offset=1,
                     alpha=255, only_draw=False, inverse=False)
```

```python
nrows = 1
ncols = 1
inv_figscale = 60
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*image_orig.shape[0]/inv_figscale,
                                                            nrows*image_orig.shape[1]/inv_figscale))
titlesize = 20

axes.imshow(doge_1)
axes.imshow(doge_2)
axes.imshow(doge_3)
axes.set_title('Sliced image', fontsize=titlesize, fontweight='bold')

plt.imsave(fname='out/doge_sliced.' + figsave_format,
           arr = doge_1 + doge_2 + doge_3,
           format=figsave_format,
           dpi=figsave_dpi)
plt.show()
```

### Depth map

```python
depth_ones = np.ones((image_orig.shape))*150
depth_map = slice_image(depth_ones, row_step=row_step, col_step=col_step, step=3, offset=1,
                        alpha=255, only_draw=True, inverse=False)
```

```python
nrows = 1
ncols = 1
inv_figscale = 60
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*depth_map.shape[0]/inv_figscale,
                                                            nrows*depth_map.shape[1]/inv_figscale))
titlesize = 20

axes.imshow(depth_map)
axes.set_title('Depth map', fontsize=titlesize, fontweight='bold')

plt.imsave(fname='out/doge_sliced_depth.' + figsave_format,
           arr = depth_map,
           format=figsave_format,
           dpi=figsave_dpi)
plt.show()
```
